```python
import json


labels = json.load(open("../dataset/ucsf-idl/labels.json"))
```

```python
page_nums = []
types = []
type2id = {}
for l in labels:
    page_nums.append(int(l["pages"]))

    t = l["type"]
    if t not in type2id:
        type2id[t] = len(type2id.items())
    types.append(type2id[t])
type2id
```

```python
print("Num samples: ", len(labels))
print("Total page count: ", sum(page_nums))
print("Max page count: ", max(page_nums))
```

```python
from matplotlib import pyplot as plt
import numpy as np

plt.hist(page_nums, bins=range(max(page_nums)+1))
```

```python
plt.hist(types, bins=range(max(types)))
```
